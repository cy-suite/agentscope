```python
import json

# Specify the path to your JSON file
file_path = "/home/users/nus/e0690420/agentscope_small_llms/examples/small_llms/ToolBenchBackbone/Llama-2-7b-chat-hf_2024-05-25_13-21-02_log_history.json"

# Open the file in read mode and load the list using json.load()
with open(file_path, 'r') as file:
    data = json.load(file)


import matplotlib.pyplot as plt
data = data[:-1]

# Extracting the 'loss' and 'step' values
steps = [item['step'] for item in data]
losses = [item['loss'] for item in data]

# Creating the plot
plt.figure(figsize=(10, 5))
plt.plot(steps, losses, marker='o')
plt.title('Loss vs Step')
plt.xlabel('Step')
plt.ylabel('Loss')
plt.grid(True)
plt.show()
```
